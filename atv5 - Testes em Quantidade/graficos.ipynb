```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Atividade 6 - Analise de Desempenho - Página1.csv")
```

```python
data.info()
```

```python
# obtendo as médias
data['média'] = data[['tempo 1', 'tempo 2']].mean(axis=1)
```

```python
data
```

## Gráfico para o teste com 10.000 mil números

```python
primeiro_teste = data[(data["entrada"] == 10000)]
```

```python
sns.barplot(primeiro_teste, x="algoritmo", y="média", palette='flare', hue="algoritmo")
plt.title("Média dos de Execução Algoritmos para o valor de 10.000 números")
plt.xlabel("Algoritmos")
plt.ylabel("Tempo de execução")
```

## Gráfico para o teste com 50.000 mil números

```python
segundo_teste = data[(data["entrada"] == 50000)]
```

```python
sns.barplot(segundo_teste, x="algoritmo", y="média", palette='flare', hue="algoritmo")
plt.title("Média de Execução dos Algoritmos para o valor de 50.000 números")
plt.xlabel("Algoritmos")
plt.ylabel("Tempo de execução")
```

## Gráfico para o teste com 100.000 mil números

```python
terceiro_teste = data[(data["entrada"] == 100000)]
```
